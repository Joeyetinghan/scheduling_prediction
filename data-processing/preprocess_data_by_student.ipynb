### sp23_post

```python
df = pd.read_excel("data_raw/2023-02-07_SP23_FinalExam_Data_Sendfile.xlsx", "Y & YO enrl", dtype={'Section': 'str', 'Catalog': 'str'})
df.head()
```

```python
df['course'] = df['Subject'] +"-"+ df['Catalog']
exam_df = df[(df['Final Exam'] == 'Y&YO') | (df['Final Exam'] == 'YO') | (df['Final Exam'] == 'Y')]
exam_df = exam_df.sort_values(by=['PS Key']).reset_index()
exam_df = exam_df[~exam_df['Exam Key'].isna()]
exam_df['Exam Key'] = exam_df['Exam Key'].astype(str)
```

```python
exam_df.head(5)
```

```python
exams = np.unique(exam_df['course'])
'CS-4780' in exams
```

```python
by_student = exam_df.groupby('ID')['course'].apply(list).reset_index()
by_student.columns = ['student', 'exams']
by_student['semester'] = 'sp23'
by_student.head()
```

```python
by_student.to_csv('./data_by_student/sp23_post.csv', header=True, index=False)
```
