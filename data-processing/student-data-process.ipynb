```python
import numpy as np
import pandas as pd
from ast import literal_eval
import math
```

```python
sp23_student_pre = pd.read_csv('./data_by_student/sp23_pre.csv')
sp23_student_pre['exams'] = sp23_student_pre['exams'].apply(literal_eval)
sp23_student_pre['semester'] = 'spring'
sp23_student_pre['year'] = 2023


fa22_student_pre = pd.read_csv('./data_by_student/fa22_pre.csv')
fa22_student_pre['exams'] = fa22_student_pre['exams'].apply(literal_eval)
fa22_student_pre['semester'] = 'fall'
fa22_student_pre['year'] = 2022


sp22_student_pre = pd.read_csv('./data_by_student/sp22_pre.csv')
sp22_student_pre['exams'] = sp22_student_pre['exams'].apply(literal_eval)
sp22_student_pre['semester'] = 'spring'
sp22_student_pre['year'] = 2022
```

```python
sp23_student_pre
```

```python
for exams in sp23_student_pre['exams']:
    for exam in exams:
        if '4780' in str(exam):
            print(exam)
```

```python
concat_student = pd.concat([sp23_student_pre, fa22_student_pre, sp22_student_pre], axis = 0)
dummy_cols = pd.get_dummies(concat_student['semester'], prefix='sem')
# concat_student = concat_student.drop('semester', axis=1)
concat_student = pd.concat([concat_student, dummy_cols], axis=1)
concat_student
```

```python
# Process information from the list of courses each student enrolled in
concat_student['num_course'] = concat_student['exams'].apply(len)
extract_departments = lambda courses: list(set([course.split('-')[0] for course in courses]))
concat_student['depts'] = concat_student['exams'].apply(extract_departments)
concat_student
```

```python
# Define a function to calculate the sum of course levels for a given list of courses
def sum_course_levels(course_list):
    course_levels = [math.floor(int(course.split('-')[1]) / 1000) for course in course_list]
    return sum(course_levels)

# Apply the sum_course_levels function to each row of the DataFrame to calculate the sum of course levels for each student
concat_student['course_level_sum'] = concat_student['exams'].apply(sum_course_levels)
concat_student
```

```python
def concatenate_lists(series):
    return list(set([item for sublist in series for item in sublist]))
# Group the DataFrame by student and semester, and concatenate the lists of courses into a dictionary
grouped_df = concat_student.groupby(['student', 'year', 'semester'])['exams'].agg(concatenate_lists)
# Display the resulting DataFrame
grouped_df = pd.DataFrame(grouped_df)
```
